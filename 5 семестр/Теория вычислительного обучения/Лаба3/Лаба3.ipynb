## Разбиение на транзакции

Далее необходимо разбить датасет на транзакции. Отличить транзакцию можно тем, что она имеет один и тот же InvoiceNo в таблице

```python
def get_transactions(df):
    dummies = pd.get_dummies(df['Description'], prefix='', prefix_sep='')
    combine = pd.concat([df[['InvoiceNo']], dummies], axis=1)
    transactions = combine.groupby('InvoiceNo').sum().reset_index().drop('InvoiceNo', axis=1)
    return transactions
```

```python
transactions = get_transactions(data)
transactions
```

Круто

## Генерация частных наборов

```python
from mlxtend.frequent_patterns import fpgrowth
```

Сгенерируем наборы

```python
frequent_itemsets = fpgrowth(transactions, min_support=0.11, use_colnames=True)
```

```python
frequent_itemsets
```

## Генерация ассоциативных правил

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
rules = rules[(rules['antecedent_len'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2)]
rules
```

```python
def view_rules(rules):
    graph = graphviz.Digraph(filename="Graph", format='png', strict=True)
    colors = get_colors()
    for _, rule in rules.iterrows():
        random_color = colors.pop(random.randrange(len(colors)))
        if len(colors) == 0:
            colors = get_colors()

        for antecedent in rule.antecedents:
            for consequent in list(rule.consequents):
                graph.edge(antecedent, consequent, color=random_color)
    return graph
```
